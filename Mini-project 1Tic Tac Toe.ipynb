```python
from IPython.display import clear_output
def display_board(board):
    print(' {s7} | {s8} | {s9} \n {s4} | {s5} | {s6} \n {s1} | {s2} | {s3} '.format(s1=board[0],s2=board[1],s3=board[2],s4=board[3],s5=board[4],s6=board[5],s7=board[6],s8=board[7],s9=board[8]))
def replay(x):
    alternate='GAME OVER'
    while alternate!=0:
        if x=='Yes':
            alternate=0
            clear_output()
        elif x=='No':
            break
        else:
            print("Please choose 'Yes' or 'No'.")
            x=input('Replay?')
            
player1 = input("Please pick a marker 'X' or 'O'")
if player1=='X':
    player2='O'
else:
    player2='X'
somelist=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
s1=somelist[0]
s2=somelist[1]
s3=somelist[2]
s4=somelist[3]
s5=somelist[4]
s6=somelist[5]
s7=somelist[6]
s8=somelist[7]
s9=somelist[8]
alternate=0

while True:
    if alternate%2==0:
        move=int(input('pick square'))
        if somelist[move-1]==' ':
            clear_output()
            somelist[move-1]=player1
            alternate+=1
            s1=somelist[0]
            s2=somelist[1]
            s3=somelist[2]
            s4=somelist[3]
            s5=somelist[4]
            s6=somelist[5]
            s7=somelist[6]
            s8=somelist[7]
            s9=somelist[8]
            if player1==s7==s8==s9 or player1==s4==s5==s6 or player1==s1==s2==s3 or player1==s7==s4==s1 or player1==s8==s5==s2 or player1==s9==s6==s3 or player1==s7==s5==s3 or player1==s9==s5==s1:
                print('Player 1 Wins! \n GAME OVER')
                display_board(somelist)
                replay(input('Replay?'))
                somelist=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
                alternate=0
            else:
                display_board(somelist)
        else:
            print('Move not possible, choose another square :)')
            continue
    else:
        move=int(input('pick square'))
        if somelist[move-1]==' ':
            clear_output()
            somelist[move-1]=player2
            alternate+=1
            s1=somelist[0]
            s2=somelist[1]
            s3=somelist[2]
            s4=somelist[3]
            s5=somelist[4]
            s6=somelist[5]
            s7=somelist[6]
            s8=somelist[7]
            s9=somelist[8]
            if player2==s7==s8==s9 or player2==s4==s5==s6 or player2==s1==s2==s3 or player2==s7==s4==s1 or player2==s8==s5==s2 or player2==s9==s6==s3 or player2==s7==s5==s3 or player2==s9==s5==s1:
                display_board(somelist)
                print('Player 2 Wins! \n GAME OVER')
                replay(input('Replay?'))
                somelist=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
                alternate=0
            else:
                display_board(somelist)
        else:
            print('Move not possible, choose another square :)')
            continue
```
